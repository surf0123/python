파이썬과 그래프

```python
import numpy as np          #(code 부분)
import matplotlib.pyplot as plt
x = np.arange(-5., 5,0.1)     # [0 1 2 3 4]
y = x ** 2              # [ 0  1  4  9 16]
z = np.sin(x)
w= x**3
plt.axis([-5,5,-3,3])
plt.plot(x,y,x,z,x,w)
plt.show()
```

막대그래프

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.bar(x, values)
plt.xticks(x, years)

plt.show()
```

파이그래프

```python
import matplotlib.pyplot as plt

ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']

plt.pie(ratio, labels=labels, autopct='%.1f%%')
plt.show()
```
