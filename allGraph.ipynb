```python
import numpy as np          
import matplotlib.pyplot as plt

x = np.arange(-5., 5, 0.1)     
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

```python
import numpy as np          
import matplotlib.pyplot as plt

x = np.arange(3)     
y =  [10, 40, 90]

plt.bar(x, y)
plt.show()
```

```python
import numpy as np          
import matplotlib.pyplot as plt

x = [10, 20, 30]
y = ['A', 'B', 'C']

plt.pie(x, labels=y, autopct='%.f')
plt.show()

```
