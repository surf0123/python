## 2. y=x^2, z =sin(x), w=x^3 그래프 그리기


```python
x = np.arange(-5., 5,0.1)     # [0 1 2 3 4]
y = x ** 2              # [ 0  1  4  9 16]
z = np.sin(x)
w= x**3
plt.axis([-5,5,-3,3])
plt.plot(x,y,x,z,x,w)
plt.show()
```

## 3. 그래프 모양 바꾸기

```python
x = np.arange(-5., 5,0.5)    
y = x ** 2              
z = np.sin(x)
w= x**3
plt.axis([-5,5,-3,3])
plt.plot(x,y,'g--',x,z,'r^-',x,w,'b*:')
plt.show()
```

## 4. 그래프 사이즈 키우기  plt.figure(figsize=(10,7))

```python
plt.figure(figsize=(10,6))
x = np.arange(-5., 5,0.5)    
y = x ** 2              
z = np.sin(x)
w= x**3
plt.axis([-5,5,-3,3])
plt.plot(x,y,'g--',x,z,'r^-',x,w,'b*:')
plt.show()
```

## 5. 축에 이름 붙이기 plt.xlabel('X-Axis')

```python

plt.figure(figsize=(10,6))
x = np.arange(-5., 5,0.5)     
y = x ** 2              
z = np.sin(x)
w= x**3
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([-5,5,-3,3])
plt.plot(x,y,'g--',x,z,'r^-',x,w,'b*:')
plt.show()
```

## 6. 제목과 범례(Legend) 표시하기

```python

plt.figure(figsize=(10,6))

x = np.arange(-5., 5,0.5)     
y = x ** 2              
z = np.sin(x)
w= x**3
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([-5,5,-3,3])
plt.plot(x,y,'g--',label='y=x^2') # 범례를 표시하기 위해 분리해야 된다
plt.plot(x,z,'r^-',label='y=sin(x)')
plt.plot(x,w,'b*:',label='y=x^3')
plt.legend()
plt.title('Graphs of Functions ')  #그래프 제목 붙이기
plt.show()
```

## 7. 그래프 제목의 글자 크기와 그래프와 거리 띄우기

```python
plt.figure(figsize=(10,6))

x = np.arange(-5., 5,0.5)     
y = x ** 2              
z = np.sin(x)
w= x**3
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([-5,5,-3,3])
plt.plot(x,y,'g--',label='y=x^2') # 범례를 표시하기 위해 분리해야 된다
plt.plot(x,z,'r^-',label='y=sin(x)')
plt.plot(x,w,'b*:',label='y=x^3')
plt.legend()

plt.rc('font', size=20)        # 기본 폰트 크기
plt.rc('axes', labelsize=20)   # x,y축 label 폰트 크기
plt.rc('xtick', labelsize=10)  # x축 눈금 폰트 크기 
plt.rc('ytick', labelsize=10)  # y축 눈금 폰트 크기
plt.rc('legend', fontsize=20)  # 범례 폰트 크기
plt.rc('figure', titlesize=50) # figure title 폰트 크기

plt.title('Graphs of Functions ',pad=20)  #그래프 제목과 그래프와 거리 띄우기
plt.show()
```

## 8. 그래프에 그리드 만들기

```python
plt.figure(figsize=(10,6))

x = np.arange(-5., 5,0.5)     
y = x ** 2              
z = np.sin(x)
w= x**3


plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([-5,5,-3,3])
plt.plot(x,y,'g--',label='y=x^2') # 범례를 표시하기 위해 분리해야 된다
plt.plot(x,z,'r^-',label='y=sin(x)')
plt.plot(x,w,'b*:',label='y=x^3')
plt.legend()

plt.rc('font', size=20)        # 기본 폰트 크기
plt.rc('axes', labelsize=20)   # x,y축 label 폰트 크기
plt.rc('xtick', labelsize=10)  # x축 눈금 폰트 크기 
plt.rc('ytick', labelsize=10)  # y축 눈금 폰트 크기
plt.rc('legend', fontsize=20)  # 범례 폰트 크기
plt.rc('figure', titlesize=50) # figure title 폰트 크기

plt.style.use('seaborn-darkgrid') #그래프에 그리드 만들기

plt.title('Graphs of Functions ',pad=20)  #그래프 제목과 그래프와 거리 띄우기
plt.show()
```
