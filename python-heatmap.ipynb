## 히트맵 그리기
matshow() 함수를 이용해서 2차원 어레이 형태의 숫자 데이터를 표현  
matshow() 함수에 어레이의 형태로 값들을 직접 입력하면 아래와 같은 그래프가 표시  

```python
import matplotlib.pyplot as plt
import numpy as np

arr = np.random.standard_normal((30, 40))

plt.matshow(arr)

plt.show()
```

컬러바 나타내기

```python
import matplotlib.pyplot as plt
import numpy as np

arr = np.random.standard_normal((30, 40))

plt.matshow(arr)
plt.colorbar()

plt.show()
```

색상 범위 지정하기  
색상의 범위를 지정하기 위해서 clim() 함수를 사용

```python
import matplotlib.pyplot as plt
import numpy as np

arr = np.random.standard_normal((30, 40))

plt.matshow(arr)
plt.colorbar(shrink=0.8, aspect=10)
plt.clim(-1.0, 1.0)
#plt.clim(-3.0, 3.0)

plt.show()
```
