## 워드 클라우드
먼저 위키피디아 모듈과 워드클라우드 모듈을 설치한다.  
Anaconda에서 Anaconda Prompt를 클릭하여 설치화면을 활성화 한 후에 다음을 입력하면 설치된다.  
pip install wikipedia  
pip install wordcloud

```python
import wikipedia

wiki = wikipedia.page('Artificial intelligence') #('Artificial intelligence')  ('America')
text = wiki.content
```

```python
from wordcloud import WordCloud
wordcloud = WordCloud(width = 2000, height =1500).generate(text)
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize =(40, 30))#40, 30
plt.imshow(wordcloud)
plt.show()
```
